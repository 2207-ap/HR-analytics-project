```python
def outliercode(df, cols, factor):
    for col in cols:
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR=Q3-Q1
        upper_whisker=Q3+(factor*IQR) #80
        lower_whisker=Q1-(factor*IQR) #25
        df[col]=np.where(df[col]>upper_whisker, upper_whisker, np.where(df[col]<lower_whisker, df[col], df[col]))
```

Note: There were no outliers on the lower side.
